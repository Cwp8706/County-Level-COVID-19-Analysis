```python
import glob
import pandas as pd
```

```python
file_list = glob.glob('**/*.csv')
```

```python
df_list = []
for file in file_list:
    df = pd.read_csv(file)
    df_list.append(df)
df = pd.DataFrame()
combined_df = df.append([df for df in df_list], ignore_index=True)
```

```python
combined_df.to_csv('pop_by_county_us_raw',index=False)
```
